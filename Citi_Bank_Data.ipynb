```python
import pandas as pd
import urllib.request
from zipfile import ZipFile 

import csv
from csv import reader

from io import StringIO
```

```python
# to make an array of dates to generate url to match Citi Bike AWS files
month = list(["01", "02", "03","04","05","06","07","08","09","10","11","12"])
year = list(["2016","2017","2018"])

i=0 # iterates through the month list
j=0 # iterates through the year list

citi_bank_data = [] # to store each dataframe to be combined
```

```python
for i in year:
    for j in month:
        
        try:  # Error Handling due to file name format inconsistencies
            
            # to create url
            url = (f"https://s3.amazonaws.com/tripdata/{i}{j}-citibike-tripdata.zip")
            
            
            #to use urlretriever to get the zip folder from Citi Bike Historical Data page
            filepath, header = urllib.request.urlretrieve(url)
            
            print(f"first try: {url}")
    
        except:
            url = (f"https://s3.amazonaws.com/tripdata/{i}{j}-citibike-tripdata.csv.zip")
            
            filepath, header = urllib.request.urlretrieve(url)
           
            print(f"second try: {url}")
        
        else: 
            
            url = url
            
            # to assign csv name
            csv = (f"{i}{j}-citibike-tripdata.csv")
            
            print(f"Gather file from: {url} & {csv}")
            
            with ZipFile(filepath, 'r') as zip:
                #extract csv file
                zip.printdir()
                data = zip.read(csv)
            
            s=str(data,'utf-8')
            data = StringIO(s)
            df= pd.read_csv(data)
            
            citi_bank_data.append(df)
```

```python
citi_bank_data_df = pd.concat(citi_bank_data,ignore_index=True, sort=False)
citi_bank_data_df.tail()
```
