```python
import pandas as pd
import urllib.request
from zipfile import ZipFile 

import csv
from csv import reader
from io import StringIO

import statistics

from sqlalchemy import create_engine


import pprint
```

### Citi_Bike_Data_Review Import Data 

A new notebook was created to maintain the data cleaning notes and results that took place.

```python
# to make an array of dates to generate url to match Citi Bike AWS files
month = list(["01", "02", "03","04","05","06","07","08","09","10","11","12"])
year = list(["2016","2017","2018"])

i=0 # iterates through the month list
j=0 # iterates through the year list

citi_bike_data = [] # to store each dataframe to be combined
```

```python
for i in year:
    for j in month:
               
        try:  # Error Handling due to file name format inconsistencies
            
            # to create url
            url = (f"https://s3.amazonaws.com/tripdata/{i}{j}-citibike-tripdata.zip")
            
            
            #to use urlretriever to get the zip folder from Citi Bike Historical Data page
            filepath, header = urllib.request.urlretrieve(url)
            
            print(f"first try: {url}")
            
            # to assign csv name
            csv = (f"{i}{j}-citibike-tripdata.csv")
            
            with ZipFile(filepath, 'r') as zip:
                    #extract csv file
                    zip.printdir()
                    data = zip.read(csv)
            
            s=str(data,'utf-8')
            data = StringIO(s)
            df= pd.read_csv(data)
            
            #to convert data column headers to lowercase 
            df.columns = map(str.lower, df.columns)
            #to rename data columns to all match 
            df.rename(columns={"tripduration":"trip duration",
                               "bikeid":"bike id",
                               "starttime":"start time",
                               "stoptime":"stop time",
                               "usertype":"user type"
                              })
            
            citi_bike_data.append(df)
            
        except:
            url = (f"https://s3.amazonaws.com/tripdata/{i}{j}-citibike-tripdata.csv.zip")
            
            filepath, header = urllib.request.urlretrieve(url)
           
            print(f"second try: {url}")
            
            
            # to assign csv name
            csv = (f"{i}{j}-citibike-tripdata.csv")
            
            with ZipFile(filepath, 'r') as zip:
                    #extract csv file
                    zip.printdir()
                    data = zip.read(csv)
            
            s=str(data,'utf-8')
            data = StringIO(s)
            df= pd.read_csv(data)
            
            #to convert data column headers to lowercase 
            df.columns = map(str.lower, df.columns)
            #to rename data columns to all match 
            df.rename(columns={"tripduration":"trip duration",
                               "bikeid":"bike id",
                               "starttime":"start time",
                               "stoptime":"stop time",
                               "usertype":"user type"
                              })

            
            citi_bike_data.append(df)
```

```python
i = 0

while i < len(citi_bike_data)-1:
    citi_bike_df = pd.concat([citi_bike_data[i],citi_bike_data[i+1]], ignore_index=True, sort=True)
    
    print("==========================")
    print(f"Completed df {i} & {i + 1} of {len(citi_bike_data)-1}")
    
    i = i + 1
```

```python
# to review final df
citi_bike_df.tail()
```

```python
from sqlalchemy import create_engine
engine = create_engine ('postgresql://postgres:postgres@localhost:5432/citi_bike_data')
citi_bike_df.to_sql('citi_bike_data_trip_history', engine)
```
