```python
import pandas as pd
import urllib.request
from zipfile import ZipFile 

import csv
from csv import reader
from io import StringIO


import statistics 

from sqlalchemy import create_engine

import pprint
```

```python
# to make an array of dates to generate url to match Citi Bike AWS files
month = list(["01", "02", "03","04","05","06","07","08","09","10","11","12"])
year = list(["2016","2017","2018"])

i=0 # iterates through the month list
j=0 # iterates through the year list

citi_bike_data = [] # to store each dataframe to be combined
```

```python
for i in year:
    for j in month:
               
        try:  # Error Handling due to file name format inconsistencies
            
            # to create url
            url = (f"https://s3.amazonaws.com/tripdata/{i}{j}-citibike-tripdata.zip")
            
            
            #to use urlretriever to get the zip folder from Citi Bike Historical Data page
            filepath, header = urllib.request.urlretrieve(url)
            
            print(f"first try: {url}")
            
            # to assign csv name
            csv = (f"{i}{j}-citibike-tripdata.csv")
            
            with ZipFile(filepath, 'r') as zip:
                    #extract csv file
                    zip.printdir()
                    data = zip.read(csv)
            
            s=str(data,'utf-8')
            data = StringIO(s)
            df= pd.read_csv(data)
            
            #to convert data column headers to lowercase 
            df.columns = map(str.lower, df.columns)
            
            citi_bike_data.append(df)
            
            
            
            
        except:
            url = (f"https://s3.amazonaws.com/tripdata/{i}{j}-citibike-tripdata.csv.zip")
            
            filepath, header = urllib.request.urlretrieve(url)
           
            print(f"second try: {url}")
            
            
            # to assign csv name
            csv = (f"{i}{j}-citibike-tripdata.csv")
            
            with ZipFile(filepath, 'r') as zip:
                    #extract csv file
                    zip.printdir()
                    data = zip.read(csv)
            
            s=str(data,'utf-8')
            data = StringIO(s)
            df= pd.read_csv(data)
            
            #to convert data column headers to lowercase 
            df.columns = map(str.lower, df.columns)
            
            citi_bike_data.append(df)
```

```python
len_of_df = len(citi_bike_data)
print(f"# of df {len_of_df}")

list_of_headers = []
list_of_columns = []

i=0

while i < len_of_df: 
    
    # to calculate the number of headers for datasets
    headers = citi_bike_data[i].columns
    # to get the value of the number of headers for datasets
    num_headers = len(headers)
    
    # to create a list of header values to evaulate what is the mode
    list_of_headers.append(headers)
    list_of_columns.append(num_headers)
       
    i += 1

print(f"dataframe header reviewed")



# to observe the number of headers vs the mode 
## if any datasets have a number of headers that do not equal the mode then the datasets
## will have an issue appeneding in the PostgreSQL database 
mode_headers = statistics.mode(list_of_columns)

variance = list(set(list_of_columns))

if mode_headers == variance[0]:
    print('same number of columns')
```

```python
# to review the column headers
## if a column header text is changed it will affect the import of the data into PostgreSQl

column_headers = {}

i=0 #iterate through dataframes
j=0 # iterate through columns 
k= i + i #to enumerate the amount of times the value occured in the dataframes


for i in range(len_of_df):
    for j in range(variance[0]):
        column_headers.update({
            list_of_headers[i][j]: (i+i)
            })
    
        print(f"Completed DataFrame {i} for column {j}")
```

```python
# to review the column headers and their frequency
sorted(column_headers.keys())
pprint.pprint(column_headers)
```

```python
## to review which columns do not match 

# the max value can also be assumed to be the mode of the keys occurance in the datasets
max_key = max(column_headers, key=column_headers.get)
max_value = column_headers[max_key]
print(max_value)

# filtering the dictionary by the max value to see which columns are new to the dataset
un_matched = {key: value for key, value in column_headers.items() if value < max_value}
print(un_matched)

# made a list of columns to rename 
rename_columns = list(un_matched.keys())

## output: ['trip duration', 'start time', 'stop time', 'bike id', 'user type']

print(rename_columns)
```
